6. Cleansing the data to make the following numerical type using apply() function:
    - Census Year, 
    - Total Population, 
    - Census Cost, 
    - Average Cost Per Person <br/>

- Census Year:

```python
df["cleansed_census_year"] = df["Census Year"].apply(lambda x : x.replace("*",""))
df["cleansed_census_year"]
df
```

- Total Population:

```python
df["cleansed_total_pop"] = df["Total Population"].apply(lambda x: x.replace(",",""))
df["cleansed_total_pop"]
df
```

- Census Cost:

```python
df["cleansed_census_cost"] = df["Census Cost"].apply(lambda x: x.replace(",","").replace("$","").replace(".","").replace(" Billion","000000000"))
df["cleansed_census_cost"]
df
```

- Average Cost Per Person:

```python
def avg_cost(cost):
    if "$" in cost:
        cost = cost.replace("$", "")
    if "," in cost:
        cost = cost.replace(",", "")
    if "cents" in cost:
        cost = cost.replace(" cents", "")
        cost = float(cost) / 100          

    return float(cost)

df["cleansed_avg_cost"] = df["Average Cost Per Person"].apply(avg_cost)
```
